### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)


tienda.head()
```

# Funciones de utilidad
Funciones que se utilizan para realizar las actividades

```python
# funciones y variables de utilidad para resolver las consignas SIN PANDAS
tiendas = [tienda, tienda2, tienda3, tienda4]

def mostrar(informe):
  # Se recibe un informe y se evalua para cada tienda de la lista
  for indice, tienda in enumerate(tiendas):
    informe(tienda, indice + 1)

def mostrar_pares(pares):
  for valor,cantidad in pares:
    print(f'{valor}: {cantidad}')

def contar_apariciones(columna):
  # Devuelve un iterable de tuplas con la (valor, cantidad)
  cantidad_por_categoria = {}
  for categoria in columna:
    if categoria in cantidad_por_categoria:
      cantidad_por_categoria[categoria] += 1
    else:
      cantidad_por_categoria[categoria] = 1
  return cantidad_por_categoria.items()

def promedio(columna):
  return sum(columna) / len(columna)

# variable para usar CON PANDAS
# Concateno los dataframes en uno solo y trabajo con ello
# pero antes agrego la columna tienda que indica a que tienda pertenece el registro
for indice, tienda in enumerate(tiendas):
  tienda['Tienda'] = indice + 1
empresa = pd.concat(tiendas)

# esta linea es para que no muestre columnas con notacion cientifica
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

#1. Análisis de facturación



```python
# SIN PANDA
# Se recibe la tienda (con indice) y se suma la columna precio
def analisis_de_facturacion(tienda, indice):
  print(f'Ingreso total de la tienda {indice} es ${sum(tienda["Precio"])}')

mostrar(analisis_de_facturacion)
```

```python
# CON PANDA
analisis_facturacion = empresa.groupby('Tienda')['Precio'].sum()
analisis_facturacion.reset_index(name='Ingreso total')
```

# 2. Ventas por categoría

```python
# SIN PANDA
def ventas_por_categoria(tienda, indice):
  cantidad_por_categoria = contar_apariciones(tienda['Categoría del Producto'])
  print(f'Tienda {indice}')
  for categoria, cantidad in cantidad_por_categoria:
    print(f'{categoria}: {cantidad}')
  print('-'*30)

mostrar(ventas_por_categoria)
```

```python
# CON PANDA
ventas_por_categoria = empresa.groupby(['Tienda','Categoría del Producto'])['Categoría del Producto'].count()
ventas_por_categoria.reset_index(name='Cantidad')
```

# 3. Calificación promedio de la tienda


```python
# SIN PANDA
# Se recibe una tienda
def calificacion_promedio(tienda, indice):
  print(f'Calificacion promedio de la tienda {indice}: {round(promedio(tienda["Calificación"]),2)}')
mostrar(calificacion_promedio)
```

```python
# CON PANDA
calificacion_promedio = empresa.groupby('Tienda')['Calificación'].mean()
calificacion_promedio.reset_index(name='Calificación promedio')
```

# 4. Productos más y menos vendidos

```python
# SIN PANDA
# Se recibe la tienda, se calcula la cantidad de cada producto
# y se ordena por la cantidad. Luego se muestra los mas y menos vendidos
def productos_vendidos(tienda, indice):
  cantidad = 3
  cantidad_por_producto = contar_apariciones(tienda['Producto'])
  print(f'Tienda {indice}')
  ranking_por_cantidad = sorted(cantidad_por_producto, key=lambda x: x[1])
  print('- Mas vendidos')
  mostrar_pares(ranking_por_cantidad[-cantidad:])
  print('- Menos vendidos')
  mostrar_pares(ranking_por_cantidad[:cantidad])
  print('-'*30)

mostrar(productos_vendidos)
```

```python
# CON PANDA
ventas_por_producto = empresa.groupby(['Tienda','Producto'])['Producto'].count().reset_index(name='Cantidad')
maximos = ventas_por_producto.groupby(['Tienda'])['Cantidad'].max().reset_index()
minimos = ventas_por_producto.groupby(['Tienda'])['Cantidad'].min().reset_index()

mas_vendidos = pd.merge(ventas_por_producto, maximos, on=['Tienda', 'Cantidad'], how='inner')
menos_vendidos = pd.merge(ventas_por_producto, minimos, on=['Tienda', 'Cantidad'], how='inner')

pd.concat([mas_vendidos, menos_vendidos]).sort_values(['Tienda', 'Cantidad'], ascending=[True, False])
```

# 5. Envío promedio por tienda

```python
# SIN PANDA
# Se recibe la tienda (con indice) y se suma la columna precio
def envio_promedio(tienda, indice):
  print(f'Envio promedio de la tienda {indice} es ${round(promedio(tienda["Costo de envío"]),2)}')

mostrar(envio_promedio)
```

```python
# CON PANDA
envio_promedio = empresa.groupby('Tienda')['Costo de envío'].mean()
envio_promedio.reset_index(name='Costo de envío promedio')
```
